## Abstract

Here, different simulations using MonteCarlo methods are presented. First, a linear fitting of random points is done several times, the fitted slope is found to follow a normal distribution. Then, these points are made to change in each iteration, so the fitted slope at the end does not show a normal distribution. The fitting is also done for three clouds of random points, with similar results to the first one. Another problem studied through MonteCarlo methods is how two non overlapping disks can be randomly distributed in a square. Also, a simulation in which distances between random points in two disks are used to get the distance between the centers is done. Finally, two non overlapping disks confined in a square are generated, but this time the disks move using a Metropolis algorithm in which the most probable position is at the center of the square, following a normal distribution.

## 1. Introduction

A given problem can be solved through random sampling. The simulations proposed here explore this way of solving problems, in no way this is intended as the most efficient method of solving these problems but is only done to illustrate MonteCarlo methods.

## 1.1 MonteCarlo

A MonteCarlo method consists in the use of random numbers to solve a problem, this is done by repetitive simulation.

## 1.2 Metropolis algorithm

The Metropolis algorithm is a MonteCarlo method which consists on iterations which seeks an equilibrium statistical solution in the following manner:

- We starty any allowed point $\textbf{r}$, so-called *walker*.

- We “shake” the point, get a new oner $\textbf{r}' = \textbf{r} + \delta$ where $\delta$ is a vector such that each of its components is a uniform random number in $[-a,a]$. This is calle "random walk".

- Evaluate $q=p(\textbf{r}')/ p(\textbf{r})$, the ratio of probabilities of new and original position.

- Accept or reject $\textbf{r}'$ by choosing a random number $t$ between 0 and 1. If t<q the new position is $\textbf{r}'$, otherwise, we keep the old position.
                                                                                         
- It eventually converges to a desired probability distribution.

## 2.4 Distance between the center of two disks

Let us assume we can find the distance between the center of two disks by measuring distances between random points contained in the disks. This is evidently wrong if both centers are in the same position, because the distance should be zero but by this method we always will get a non-zero result. The case of two very far away disks is also of no interest because then the diks could be considered points and the result would clearly be approximately correct.

In this simulation sets of 1000 random points contained inside two diks of radii 1 are generated. One disk has center at (0,0) and the other has it's center at (1.5,1.5) as shown inf Figure 8:



<center>Figure 8: One disk at (0,0) and another at (1.5,1.5). 1000 random points contained in each.</center>

The distance between both centers is then 2.1213, that is, 1.0607 times the diameter of the disks. It would then be interesting if the assumption described above holds and we get an answer close to the distance between the centers.

After generating and measuring distances between two random points, one in each disk, 10000 times, we get the histogram shown in figure 9 which fits very well to a normal distribution (red line).



<center>Figure 9: Histograms of distances between random points in the disks.</center>

The mean obtained is 2.2422 which compared to the expected distance 2.1213 gives a 5.70% error.

## 3. Conclusions

The simulations presented here used random sampling to get results. If each iteration in a simulation uses a different set of random points, then the results differ but if the method of random generation of data is consistent, then the results do vary following a symmetric distribution, as seen in most of the histograms, presented this is usually a Gaussian distribution. If the sampling changes with time, then a non-normal distribution can be obtained, even a non-symmetric one as in Figure 3.

These simulations also illustrate how a MonteCarlo method can succesfully give approximate results to a given problem.

```r
steps=1000
X <- rep(NA, steps)
for (i in 1:steps) {

N=100
# x is random normal with arbitrary mean. 
# y is random normal with mean x, in this way we get something like a linear function.
xa=rnorm(N, mean=-1, sd=0.5)
ya=rnorm(N, mean=xa, sd=0.5)

xb=rnorm(N, mean=0, sd=0.5)
yb=rnorm(N, mean=xb, sd=0.5)

xc=rnorm(N, mean=1, sd=0.5)
yc=rnorm(N, mean=xc, sd=0.5)


y_fit<-ya+yb+yc
x_fit<-xa+xb+xc
constant_model <- lm(y_fit ~ x_fit )
X[i]<-constant_model$coefficients[2]


}
plot(xa,ya, xlim = range(c(-2,2)) ,ylim = range(c(-2,2)) )
points(xb,yb)
points(xc,yc)
abline(constant_model, lwd = 3, col = "red")
print(constant_model$coefficients[2])
hist(X, freq=FALSE)
curve(dnorm(x,mean=mean(X),sd=sd(X)),col="red",add=T)
print(mean(X) )
```

```r
### Creating 2 random disks of radii R inside a square of side L several times, just filling the space.
    
L=1 #square side
R=0.25  #circles radii #The limit for square of side 1 would be R=0.585786/2. Ref:https://mathworld.wolfram.com/CirclePacking.html. Beware, if you get close to the limit, the code will be slow.

xa=NA
ya=NA
plot(xa,ya, xlim = range(c(0,L)) ,ylim = range(c(0,L)))
steps=1000

XA <- rep(NA, steps)
YA <- rep(NA, steps)
XB <- rep(NA, steps)
YB <- rep(NA, steps)
for (i in 1:steps) 
  {

    #First disk
    xa=runif(1,R,L-R)
    ya=runif(1,R,L-R)

    #Second disk
    xb=runif(1,R,L-R)
    yb=runif(1,R,L-R)

    #We want both disks to be more than 2 times R apart. If this is not true, we keep generating random positions.
    while ( sqrt( (xa - xb)^2 + (ya - yb)^2 ) < 2*R ) {
    xa=runif(1,R,L-R)
    ya=runif(1,R,L-R)
    xb=runif(1,R,L-R)
    yb=runif(1,R,L-R)
    }

    XA[i]=xa
    YA[i]=ya
    XB[i]=xb
    YB[i]=yb
    #Plotting
    points(xa,ya, xlim = range(c(0,L)) ,ylim = range(c(0,L)), pch = 17 )
    points(xb,yb, pch = 15 )
  }

hist(XA)
hist(YA)
hist(XB)
hist(YB)
```

```r
#Random points cointained in two disks of same radii. The positions of disks can be changed.
##Distances between random points in the disks are calculated. 
##The further away the disks are, the mean of these distances approximate the true distance.

steps=10000
r <- rep(NA, steps)
for (i in 1:steps) {

N=1000
R=1
cxa=0
cya=0
cxb=1.5
cyb=1.5

xa=runif(N,-R+cxa,R+cxa)
ya=runif(N,-R+cya,R+cya)
xb=runif(N,-R+cxb,R+cxb)
yb=runif(N,-R+cyb,R+cyb)


xca = ifelse( sqrt( (xa-cxa)^2 + (ya-cya)^2 ) > R, NA, xa)
yca = ifelse( sqrt( (xa-cxa)^2 + (ya-cya)^2 ) > R, NA, ya)
xcb = ifelse( sqrt( (xb-cxb)^2 + (yb-cyb)^2 ) > R, NA, xb)
ycb = ifelse( sqrt( (xb-cxb)^2 + (yb-cyb)^2 ) > R, NA, yb)

distance=sqrt( (xca - xcb)^2 + (yca - ycb)^2)
r[i]=mean(distance, na.rm = TRUE)

}

low=min(xa,xb,ya,yb)
high=max(xa,xb,ya,yb)

plot(xca,yca,xlim = range(c(low,high)) ,ylim = range(c(low,high))  )
points(xcb,ycb)

hist(r, freq=FALSE)
curve(dnorm(x,mean=mean(r),sd=sd(r)),col="red",add=T)
print(mean(r) )


expected=sqrt( (cxa-cxb)^2 + (cya-cyb)^2 )
paste("mean distance by algorithm",mean(r) )
paste("expected distance",expected)
error=( mean(r)-expected)/expected
paste("error is",error)


```

```r
### Creating 2 random disks of radii R inside a square of side L. These disk seek the center through normal dist.

L=1 #square side
R=0.1  #circles radii

#Uniform random number, from R to L-R
ran<-function(){runif(1,R,L-R)}

#Uniform random number, from -R to +R
delran<-function(){runif(1,-R,R)}

#First disk
xa=ran()
ya=ran()

#Second disk
xb=ran()
yb=ran()

#We want both disks to be more than 2 times R apart. If this is not true, we keep generating random positions.
while ( sqrt( (xa - xb)^2 + (ya - yb)^2 ) < 2*R ) {
    xb=ran()
    yb=ran()
}

#Plotting
plot(xa,ya, xlim = range(c(0,L)) ,ylim = range(c(0,L)))
points(xb,yb)

###From here on, Metropolis Algorithm


del=1
   
    
normal_prob<-function(x){dnorm(x, mean = L/2, sd = 0.1, log = FALSE)}
#normal_prob(L/2)



steps=1000
XA <- rep(NA, steps)
YA <- rep(NA, steps)
XB <- rep(NA, steps)
YB <- rep(NA, steps)

for (i in 1:steps) 
  {
    xa_temp=xa
    ya_temp=ya
    xb_temp=xa #This is to make sure, a while bellow works from the first step.
    yb_temp=ya
    while ( sqrt( (xa_temp - xb_temp)^2 + (ya_temp - yb_temp)^2 ) < 2*R ) {
    xa_walk<-xa+del*delran()
    ya_walk<-ya+del*delran()
    xb_walk<-xb+del*delran()
    yb_walk<-yb+del*delran()

    U=runif(1,0,1)

    xa_temp=ifelse(U<normal_prob(xa_walk)/normal_prob(xa),xa_walk,xa)
    ya_temp=ifelse(U<normal_prob(ya_walk)/normal_prob(ya),ya_walk,ya)
    xb_temp=ifelse(U<normal_prob(xb_walk)/normal_prob(xb),xb_walk,xb)
    yb_temp=ifelse(U<normal_prob(yb_walk)/normal_prob(yb),yb_walk,yb)

    xa=xa_temp
    ya=ya_temp
    xb=xb_temp
    yb=yb_temp
    XA[i]=xa
    YA[i]=ya
    XB[i]=xb
    YB[i]=yb
    points(xa,ya, col= rgb(i/steps, 0, 1-i/steps), pch = 17 )
    points(xb,yb, col= rgb(i/steps, 0, 1-i/steps), pch = 15 )
    #print(xa)
    }

  }

hist(XA)
hist(YA)
hist(XB)
hist(YB)
```
